```python
fighters_stats = pd.DataFrame()
fighters_stats["fighter"] = pd.concat([stats["red_fighter_name"], stats["blue_fighter_name"]])
fighters_stats["result"] = pd.concat([stats["red_fighter_result"], stats["blue_fighter_result"]])
fighters_stats["nickname"] = pd.concat([stats["red_fighter_nickname"], stats["blue_fighter_nickname"]])
fighters_stats["event_date"] = stats["event_date"].reset_index(drop=True)
```

```python
# find match for fighters and fighters_stats
fighter_names = fighters["fighter_name"].unique()
fighter_stats_names = fighters_stats["fighter"].unique()
```

```python
fighter_names = [str(name).lower() for name in fighter_names]
fighter_stats_names = [str(name).lower() for name in fighter_stats_names]
```
